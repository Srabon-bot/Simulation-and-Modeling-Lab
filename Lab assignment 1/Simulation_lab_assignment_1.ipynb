```python
# NumPy is used for matrix and numerical operations
import numpy as np

# Matplotlib is used for plotting
import matplotlib.pyplot as plt
```

```python
# Creating two 3x3 matrices with random integers between 1 and 9
matrix_A = np.random.randint(1, 10, (3, 3))
matrix_B = np.random.randint(1, 10, (3, 3))

print("Matrix A:")
print(matrix_A)

print("\nMatrix B:")
print(matrix_B)
```

```python
# Matrix addition
addition = matrix_A + matrix_B

# Matrix subtraction
subtraction = matrix_A - matrix_B

# Matrix multiplication
multiplication = np.dot(matrix_A, matrix_B)

print("\nAddition of A and B:")
print(addition)

print("\nSubtraction of A and B:")
print(subtraction)

print("\nMultiplication of A and B:")
print(multiplication)
```

```python
# Calculating determinant of Matrix A
determinant = np.linalg.det(matrix_A)

# Calculating rank of Matrix A
rank = np.linalg.matrix_rank(matrix_A)

print("\nDeterminant of Matrix A:", determinant)
print("Rank of Matrix A:", rank)

# Checking if matrix is invertible
if determinant != 0:
    inverse = np.linalg.inv(matrix_A)
    print("\nInverse of Matrix A:")
    print(inverse)
else:
    print("\nMatrix A does not have an inverse (determinant is zero).")
```

```python
# Creating two vectors with 15 random float values
vector_1 = np.random.rand(15)
vector_2 = np.random.rand(15)
```

```python
# Plotting both vectors on the same graph
plt.plot(vector_1, label="Vector 1", marker='o')
plt.plot(vector_2, label="Vector 2", marker='x')

# Labeling axes and title
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Comparison of Two Random Vectors")

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
