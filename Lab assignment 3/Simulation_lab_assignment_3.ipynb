```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Part 1
# Set a random seed for reproducibility
np.random.seed(42)

# Task 1: Generate 3,000 wait times — Uniform Distribution (Regular Days)
# Uniform distribution between 2 and 12 minutes
regular_wait = np.random.uniform(low=2, high=12, size=3000)

print("Regular Day (Uniform) — First 10 values:")
print(np.round(regular_wait[:10], 2))
print(f"Total samples generated: {len(regular_wait)}")
```

```python
# Task 2: Generate 3,000 wait times — Normal Distribution (Busy Days)
# Mean = 8, Standard Deviation = 2
busy_wait = np.random.normal(loc=8, scale=2, size=3000)

print("Busy Day (Normal) — First 10 values (before cleaning):")
print(np.round(busy_wait[:10], 2))
```

```python
# Task 3: Ensure no negative wait times — Replace negatives with 0
negative_count = np.sum(busy_wait < 0)
print(f"Number of negative values before cleaning: {negative_count}")

busy_wait = np.where(busy_wait < 0, 0, busy_wait)

print(f"Number of negative values after cleaning: {np.sum(busy_wait < 0)}")
print("Busy Day (Normal) — First 10 values (after cleaning):")
print(np.round(busy_wait[:10], 2))
```

```python
#Part 2
# Plot 1: Two Subplots — Side by Side Histograms
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Subplot 1 — Regular Days (Uniform)
axes[0].hist(regular_wait, bins=30, color='steelblue', edgecolor='black', alpha=0.85)
axes[0].set_title('Regular Days\n(Uniform Distribution)', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Wait Time (minutes)', fontsize=11)
axes[0].set_ylabel('Frequency', fontsize=11)
axes[0].grid(axis='y', linestyle='--', alpha=0.6)

# Subplot 2 — Busy Days (Normal)
axes[1].hist(busy_wait, bins=30, color='tomato', edgecolor='black', alpha=0.85)
axes[1].set_title('Busy Days\n(Normal Distribution)', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Wait Time (minutes)', fontsize=11)
axes[1].set_ylabel('Frequency', fontsize=11)
axes[1].grid(axis='y', linestyle='--', alpha=0.6)

plt.suptitle('Customer Wait Time Simulation — Regular vs Busy Days', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Overlaid Histograms with Transparency
plt.figure(figsize=(10, 5))

plt.hist(regular_wait, bins=30, color='steelblue', edgecolor='black', alpha=0.55, label='Regular Days (Uniform)')
plt.hist(busy_wait,    bins=30, color='tomato',    edgecolor='black', alpha=0.55, label='Busy Days (Normal)')

plt.title('Overlaid Histogram — Customer Wait Times', fontsize=13, fontweight='bold')
plt.xlabel('Wait Time (minutes)', fontsize=11)
plt.ylabel('Frequency', fontsize=11)
plt.legend(fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
#Part 3
# Calculate statistics for Regular Days (Uniform)
reg_mean   = np.mean(regular_wait)
reg_median = np.median(regular_wait)
reg_std    = np.std(regular_wait)
reg_max    = np.max(regular_wait)

# Calculate statistics for Busy Days (Normal)
busy_mean   = np.mean(busy_wait)
busy_median = np.median(busy_wait)
busy_std    = np.std(busy_wait)
busy_max    = np.max(busy_wait)

# Display results in a formatted table
print("=" * 52)
print(f"{'Statistical Analysis':^52}")
print("=" * 52)
print(f"{'Metric':<20} {'Regular Days':>15} {'Busy Days':>15}")
print("-" * 52)
print(f"{'Mean (minutes)':<20} {reg_mean:>15.4f} {busy_mean:>15.4f}")
print(f"{'Median (minutes)':<20} {reg_median:>15.4f} {busy_median:>15.4f}")
print(f"{'Std Deviation':<20} {reg_std:>15.4f} {busy_std:>15.4f}")
print(f"{'Max Wait Time':<20} {reg_max:>15.4f} {busy_max:>15.4f}")
print("=" * 52)
```
